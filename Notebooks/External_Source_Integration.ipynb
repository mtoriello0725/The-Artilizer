## Songkick - Search for an artist

https://api.songkick.com/api/3.0/search/artists.json?apikey={your_api_key}&query={artist_name}

### Search for an Artist: Songkick requires user to have artist_id

```python
from config import SK_API_KEY

# URL : https://api.songkick.com/api/3.0/search/artists.json?apikey={your_api_key}&query={artist_name}

url = 'https://api.songkick.com/api/3.0/'
method = 'search/artists.json?apikey='
API_KEY = SK_API_KEY
query = '&query='
artist = 'Foster The People'

query_url = url + method + API_KEY + query + artist
```

```python
search_json_response = requests.get(query_url).json()
search_json_response
```

```python
ID = search_json_response['resultsPage']['results']['artist'][0]['id']
ID
```

### Once artist_id is obtained, find the following data: 
* Past Concerts
* Future Concerts

#### Past Concerts

```python
# https://api.songkick.com/api/3.0/artists/{artist_id}/gigography.json?apikey={your_api_key}

query_url_2 = url + 'artists/' + str(ID) + '/gigography.json?' + 'order=desc' + '&apikey=' + API_KEY
query_url_2
```

```python
artist_json_response = requests.get(query_url_2).json()
artist_json_response
```

```python
events = artist_json_response['resultsPage']['results']['event']
```

```python
date = []

for i in range (0,len(events)):
    date.append(events[i]['start']['date'])

date
date[-1]

# only receiving 50 entries... Total Entries is on bottom of resultsPage
```

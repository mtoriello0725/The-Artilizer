# Pull data from SQLITE

### Configure necessary imports

```python
#Necessary imports
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
from sqlalchemy import func
```

```python
# Create Engine
engine = create_engine("sqlite:///../Resources/artists.sqlite", echo=False)
```

```python
conn = engine.connect()
```

```python
import numpy as np
import pandas as pd
```

```python
ledZep_df = pd.read_sql_table("Led Zeppelin", con=conn)
```

```python
attr_list = "acousticness, danceability, energy, instrumentalness, liveness, speechiness, valence"
haken_df = pd.read_sql_query(f"SELECT {attr_list} FROM Haken", con=conn)
```
